```python
# Loading the packages 
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Basic Barplot
# Make a dataset:
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')

# Create bars
plt.bar(bars, height)
 
# Create names on the x-axis
plt.xticks(bars)
 
# Show graphic
plt.show()
```

```python
#cloumn chart 
# Create horizontal bars
plt.barh(bars, height)
 
# Create names on the y-axis
plt.yticks(bars)
 
# Show graphic
plt.show()
```

```python
#Adjusting the colors

# RGB is a way of making colors.
# color =  red, green and blue + the transparency
# Put transparency = 0 
# Put transparency = 1
plt.bar(bars, height, color=(0.2, 0.4, 0.6, 0.6))
plt.xticks(bars)
plt.show()
```

```python
# Different color for each bar 

# Provide a list of color names to the color argument
plt.bar(bars, height, color=['black', 'red', 'green', 'blue', 'cyan'])
plt.xticks(bars)
plt.show()
```

```python
# Control color of the border 


# The edgecolor argument allows to color the borders of barplots.
plt.bar(bars, height, color=(0.1, 0.1, 0.1, 0.1),  edgecolor='blue')
plt.xticks(bars)
plt.show()
```

```python
#Add title and axis labels

# Create bars and choose color
plt.bar(bars, height, color = (0.5,0.1,0.5,0.6))
 
# Add title and axis names
plt.title('$Height_{bars}$', fontsize=14)
plt.xlabel('Name', fontsize=14)
plt.ylabel('Height', fontsize=14)
 
# Limits for the Y axis
plt.ylim(0,60)
 
# Create names
plt.xticks(bars)
 
# Show graphic
plt.show()
```

```python
#Basic Line Chart

# Create a dataframe 
df = pd.DataFrame({'year': [2015, 2016, 2017, 2018, 2019], 'T20': [24, 36, 12, 18, 25], 'ODI': [89, 24, 37, 63, 98]})

# Draw the plot 
plt.plot( 'year', 'T20', data = df)
plt.show()
```

```python
# Draw the plot 

plt.plot( 'year', 'T20', data = df, linewidth = 5)
plt.plot('year', 'ODI', data = df, c='red', linewidth = 5)
plt.xticks(np.arange(min(df.year), max(df.year)+1, 1.0), fontsize = 15)
plt.yticks(np.arange(0,125,25), fontsize = 15)
plt.title('Points scored by India', fontsize = 15)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Points', fontsize = 15)
plt.legend(loc='upper center')
plt.grid(b=True, alpha=0.9) #which argument to choose the axis 
plt.show()
```

```python
#Plotting the random values

# create data
values=np.cumsum(np.random.randn(1000,1))
# use the plot function
plt.plot(values)
plt.show()
```

```python
#Multiple lines chart

# Data
df=pd.DataFrame({'x': range(1,11), 'y1': np.random.randn(10), 'y2': np.random.randn(10)+range(1,11), 'y3': np.random.randn(10)+range(11,21) })
 
# multiple line plot
plt.plot( 'x', 'y1', data=df, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot( 'x', 'y2', data=df, marker='', color='olive', linewidth=2)
plt.plot( 'x', 'y3', data=df, marker='', color='olive', linewidth=2, linestyle='dashed', label="toto")
plt.legend(loc='upper center')
plt.show()
```

```python
movies = pd.read_csv('movies.csv')
movies = movies.drop(['Unnamed: 0'], axis = 1)
movies.head()
```

```python
#scatter plot

sns.regplot(x=movies["critics_score"], 
            y = movies["audience_score"], 
            fit_reg=False)
plt.show()
# Without regression fit:
```

```python
# use the function regplot to make a scatterplot with regression line 
sns.set(color_codes=True) 
sns.regplot(x = movies["critics_score"], 
            y = movies["audience_score"], ci = 95) 
# ci(confidence interval) = None will not displayany interval
```

```python
# Another example to show the confidence interval

tips = sns.load_dataset("tips")
# tips.head()
sns.regplot(x="total_bill", y="tip", data=tips, ci = None)
# sns.regplot(x="total_bill", y="tip", data=tips)
# sns.regplot(x="total_bill", y="tip", data=tips, ci = 95)
```

```python
# Customizing the scatter plot

sns.regplot(x = movies["critics_score"], 
            y = movies["audience_score"], 
            marker="+", fit_reg=False)
```

```python
sns.regplot(x = movies["critics_score"], 
            y = movies["audience_score"], 
            marker="+", fit_reg=False, color='red')
```

```python
sns.regplot(x = movies["critics_score"], 
            y = movies["audience_score"], 
            marker="+", fit_reg=False, scatter_kws={"color":"red", "alpha":0.5, "s":100})
# alpha - obesity of marker and s-size of the marker
```

```python
# Using categorical variable for color
# reg plot is subset of lm plot. 
# In 'lm plot' we can add hue as additional feature.

# Use the 'hue' argument to provide a factor variable
sns.lmplot(x = "critics_score", 
            y = "audience_score", data = movies, fit_reg=False, hue='genre', legend=False)
 
# Move the legend (labeling of plot) to an empty part of the plot
plt.legend(loc='lower right')
```

```python
# Another example 

df = sns.load_dataset('iris')
 
# Use the 'hue' argument to provide a factor variable
sns.lmplot(x="sepal_length", y="sepal_width", data=df, fit_reg=False, hue='species', legend=False)
 
# Move the legend to an empty part of the plot
plt.legend(loc='lower right')
```

```python
#visualizing distribution

# Violin plot

# https://python-graph-gallery.com/violin-plot/
# Violinplots allow to visualize the distribution of a numeric variable for one or several groups. 
# It is really close from a boxplot, but allows a deeper understanding of the density. 
# Violins are particularly adapted when the amount of data is huge and showing individual observations gets impossible. 
# Violinplots probably deserve more attention compared to boxplot that can sometimes hide features of the data.

# If you have only one numerical variable, you probably better have to make an histogram or a density plot. 
# But you can still use the violinplot function to describe the distribution of this variable, as follow:
# from pylab import *
df = sns.load_dataset('iris')
 
# plot
sns.violinplot(x=df["species"], y=df["sepal_length"])
# sns.boxplot(x="species", y="sepal_length", data=df)
```

```python
sns.violinplot( x=df["species"], y=df["sepal_length"], palette="Blues")
```

```python
# Make a dictionary with one specific color per group:
my_pal = {"versicolor": "g", "setosa": "b", "virginica":"m"}
 
#plot it
sns.violinplot( x=df["species"], y=df["sepal_length"], palette=my_pal)
```

```python
# Density plot

# plot of 2 variables
# Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable
p1=sns.kdeplot(df['sepal_width'], shade=True, color="r")
p1=sns.kdeplot(df['sepal_length'], shade=True, color="b")
```

```python
from pylab import * 
df = sns.load_dataset('iris')
subplot(2,2,1)
# density plot with shade
sns.kdeplot(df['sepal_width'], shade=True, color='r')

subplot(2,2,2)
# density plot with shade
sns.kdeplot(df['sepal_length'], shade=True, color='b')

subplot(2,2,3)
# density plot with shade
sns.kdeplot(df['petal_width'], shade=True, color='g')

subplot(2,2,4)
# density plot with shade
sns.kdeplot(df['petal_length'], shade=True, color='y')
```

```python
#Unimodal Distributions. - sepal_width
#Unimodal distributions have a single peak, or mode.Normal distribution (Bell curve) is unimodal.


#Bimodal Distribution for petal width and petal length
#Youve got two peaks of data, which usually indicates youve got two different groups. The bi in bimodal distribution 
#refers to two and modal refers to the peaks.


#Multimodal Distributions
#Multimidal distributions have more than two peaks.
#f you cant clearly find one peak or two peaks in a graph, the likelihood is that you either have a uniform 
#distribution (where all the peaks are the same height) or a multimodal distribution, where there are several peaks of the same height.


# Fun fact: While the bell curve is normally associated with grades (i.e. 5% of the class will get an A and 10% of the class will get a B), 
# its also quite normal to have a bimodal distribution where roughly half of a class will do very well (getting As and Bs) and the other half 
# of the class will receive poor grades (Ds and Fs). Bimodal distributions are very common in college freshman mathematics classes!
```

```python
subplot(2,2,1)
# density plot with shade
sns.kdeplot(df['sepal_width'], shade=True, color='r', cumulative=True)

subplot(2,2,2)
# density plot with shade
sns.kdeplot(df['sepal_length'], shade=True, color='b', cumulative=True)

subplot(2,2,3)
# density plot with shade
sns.kdeplot(df['petal_width'], shade=True, color='g', cumulative=True)

subplot(2,2,4)
# density plot with shade
sns.kdeplot(df['petal_length'], shade=True, color='y', cumulative=True)
```
